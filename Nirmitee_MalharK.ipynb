```python
import unicodedata

# There are various symobls in the text.
# Took the range of such char, and compared it with each char and accordingly appended clean char and later joined them into a single string.
def remove_pua_symbols(text):
    cleaned_text = []
    for char in text:
        if ord(char) < 0xE000 or ord(char) > 0xF8FF:
            cleaned_text.append(char)
    return ''.join(cleaned_text)

# Apply the function to 'Content' column
df['Content'] = df['Content'].apply(remove_pua_symbols)
```

```python
#Used tf-idf vectorizer to convert it into vectors, removing common english stopwards.
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(df['Content'])
```

```python
job_description = """We are looking for an experienced AI Engineer for a
client project which is specializing in fraud detection using Large Language Models (LLMs) and traditional
machine learning models. This role will involve
designing and developing cutting-edge AI solutions for
client projects while contributing to future-focused
prediction and insights products. The role will evolve to
work in other domains including healthcare where ever
insights and decision making can be done using AI.
Collaborate with cross-functional teams to
understand client requirements and translate
them into scalable AI models.
Design, develop, and implement fraud detection
models using LLMs and traditional machine
learning techniques.
Build and maintain data pipelines for training,
testing, and deploying models efficiently.
Develop prediction and insights tools to provide
actionable intelligence for fraud prevention.
Python, LangChain, LlamaIndex,
pandas, ML algorithms, AI algorithms, Fine
Tuning, REST APIS

"""
```

```python
job_desc_vector = vectorizer.transform([job_description.lower()]) #Converted JD to vectors for comparing similarity.
```

```python
# Calculated similarity scores
from sklearn.metrics.pairwise import cosine_similarity
df['similarity'] = cosine_similarity(X, job_desc_vector).flatten()

# Ranked resumes based on similarity scores
df = df.sort_values(by='similarity', ascending=False)

# Displayed top 5 ranked resumes
df[['name', 'similarity']].head()
```

```python
#Defined a function to do the same things as above for the user input.
def get_relevant_resumes(job_description,df):

  job_desc_vector = vectorizer.transform([job_description.lower()])

  df['similarity'] = cosine_similarity(X, job_desc_vector).flatten()

  df = df.sort_values(by='similarity', ascending=False)

  return df[['name', 'similarity']].head()
```

```python
# User Input for the JD
job_description = input("Enter the job description: ")

relevant_resumes = get_relevant_resumes(job_description,df)

#Printing the relevant resumes
print("Top 5 relevant resumes:")
print(relevant_resumes)
```
